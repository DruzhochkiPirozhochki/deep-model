```python
import os
from PIL import Image
import numpy as np
```

```python
PATH_FULL_IMAGE = "./our_data/"
PATH_NEW_DATASET_TXT = "./cut_imgs/"
```

```python
files = os.listdir(PATH_FULL_IMAGE)

for file_name in files:
        if file_name.endswith('.txt'):
            with open(PATH_FULL_IMAGE + file_name) as f:
                lines = f.read().splitlines()
                
                pillar = np.array(list(map(float, [line for line in lines if line.startswith("2")][0].split(' ')[1:])))
                shift = np.concatenate((0.5 - pillar[:2], np.array([0, 0])))
                scale = np.tile(1 / pillar[2:], 2)
                
                insulators = [line for line in lines if line.startswith("1")]
                bumpers = [line for line in lines if line.startswith("0")]
                new_insulators = []
                print(file_name)
                for insulator in insulators:
                    coords = np.array(list(map(float, insulator.split()[1:])))
                    new_coords = (coords-np.array([0.5, 0.5, 0, 0]) + shift) * scale + np.array([0.5, 0.5, 0, 0])
                    if not(new_coords[0] <= 0 or new_coords[0] >=1 or new_coords[1] <= 0 or new_coords[1] >= 1):
                        new_insulators.append(" ".join(list(map(str, ["0"] + new_coords.tolist()))))
                for bumper in bumpers:
                    coords = np.array(list(map(float, bumper.split()[1:])))
                    new_coords = (coords-np.array([0.5, 0.5, 0, 0]) + shift) * scale + np.array([0.5, 0.5, 0, 0])
                    if not(new_coords[0] <= 0 or new_coords[0] >=1 or new_coords[1] <= 0 or new_coords[1] >= 1):
                        new_insulators.append(" ".join(list(map(str, ["1"] + new_coords.tolist()))))
            with open(PATH_NEW_DATASET_TXT + file_name.rsplit(".", 1)[0] + "_pillar_0.txt", 'w') as f:
                f.write("\n".join(new_insulators))
```

```python
import os
from PIL import Image

PATH = './cut_imgs/pillar/'


def crop_images():
    
    os.mkdir(PATH + 'reshaped')
    files = os.listdir(PATH)

    for file_name in files:
        if file_name.endswith('.jpg'):
            image = Image.open(PATH + file_name)
            new_img = image.resize((416, 416))
            new_img.save(PATH + 'reshaped/' + file_name[:-4] + '.jpg')
```

```python
crop_images()
```
